```python
# Function to create a bar plot with a colormap
def create_bar_plot(data, ylabel, title):
    plt.figure(figsize=(12, 8))
    
    # Extracting values and genres
    values = data['mean']
    genres = data.index

    # Sorting values in descending order
    data_sorted = data.sort_values(by='mean', ascending=False)
    
    # Creating a colormap
    colors = plt.cm.viridis(np.linspace(0, 1, len(values)))

    # Plotting the bar chart
    bars = plt.bar(data_sorted.index, data_sorted['mean'], color=colors)

    # Adding a colorbar for reference
    cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis))
    cbar.set_label('Scale')

    # Adding labels and title
    plt.title(title)
    plt.xlabel('Genre')
    plt.ylabel(ylabel)
    plt.xticks(rotation=45, ha='right')

    # Adding value annotations on top of each bar with adjusted positions and formatting
    for bar, value in zip(bars, data_sorted['mean']):
        rounded_value = round(value / 1000) * 1000  # Round to nearest thousand
        plt.text(bar.get_x() + bar.get_width() / 2, value + 0.05, f'${rounded_value:,.0f}', ha='center', va='bottom', fontsize=8)

    plt.show()

# Specify the genres you want to include
selected_genres = ['Horror', 'Mystery', 'Thriller', 'Sport', 'Romance', 'Music', 'Sci-Fi', 'Documentary', 'Animation', 'Comedy']

# Exclude the 'Missing' genre from the DataFrame
drop_dup_df_genre_filtered = drop_dup_df_genre[drop_dup_df_genre['genres'] != 'Missing']

# Mean Production Budget by Genre for selected genres
production_budget_stats_selected_genres = drop_dup_df_genre_filtered[drop_dup_df_genre_filtered['genres'].isin(selected_genres)].groupby('genres')['production_budget'].agg(['mean', 'median', 'count'])
print("Mean Production Budget by Genre for Selected Genres:")
print(production_budget_stats_selected_genres)

# Create bar plot for mean production budget for selected genres
create_bar_plot(production_budget_stats_selected_genres[['mean']], 'Mean Production Budget', 'Mean Production Budget by Selected Genres')
```

```python
# Function to create a bar plot with a colormap and customized text color
def create_bar_plot(data, ylabel, title, text_color_dict):
    plt.figure(figsize=(12, 8))
    
    # Extracting values and genres
    values = data['mean']
    genres = data.index

    # Rounding values and formatting as integers
    rounded_values = np.round(values, decimals=-3).astype(int)

    # Creating a colormap
    colors = plt.cm.viridis(np.linspace(0, 1, len(values)))

    # Plotting the bar chart
    bars = plt.bar(genres, values, color=colors)

    # Adding a colorbar for reference
    cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis))
    cbar.set_label('Scale')

    # Adding labels and title
    plt.title(title)
    plt.xlabel('Genre')
    plt.ylabel(ylabel)

    # Rotating x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Adding value annotations with rotated text at the center of each bar
    for bar, rounded_value, genre in zip(bars, rounded_values, genres):
        color = text_color_dict.get(genre, 'black')  # Default to black if genre not in the dictionary
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() / 2,  # Adjust the position to the center of the bar
            f'{rounded_value:,}',  # Use commas for thousands separator
            ha='center',
            va='center',  # Use 'center' for vertical alignment
            rotation='vertical',  # Rotate text vertically
            color=color  # Set the text color based on the genre
        )

    # Autoscale y-axis ticks
    plt.autoscale()

    plt.show()

# Define a dictionary mapping genres to text colors
text_color_dict = {
    'Animation': 'white',
    'Sci-Fi': 'white',
    'Comedy': 'white',
    'Sport': 'white',
    'Thriller': 'white',
    'Mystery': 'black',
    'Music': 'black',
    'Horror': 'black',
    'Romance': 'black',
    'Documentary': 'black'
}

# Create bar plot with customized text colors
create_bar_plot(top_ten_domestic_gross[['mean']], 'Mean Domestic Gross', 'Top Ten Mean Domestic Gross by Genre', text_color_dict)
```
